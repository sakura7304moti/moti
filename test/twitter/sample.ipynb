```python
import pandas as pd
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
import time
import os
import urllib
import datetime
```

```python
scraper = sntwitter.TwitterSearchScraper("#絵かゆ")
```

```python
today = datetime.datetime.today()
```

```python
tweets = []
for index,tweet in enumerate(scraper.get_items()):
    images = []
    if tweet.likeCount > 4:
        try:
            for media in tweet.media:
                images.append(media.fullUrl)

            data = [
                tweet.url,
                tweet.date,
                images
            ]
            tweets.append(data)
        except:
            pass
    rep_date = tweet.date.replace(tzinfo=None)
    if (today - rep_date).days >= 7:
        break
```

```python
tweet_df = pd.DataFrame(
    tweets,columns = ["url","date","images"]
)
```

```python
def download(url):
    file_name = url.split('/')[-1].split('?')[0]+'.jpg'
    dir_name = r'E:\Project\moti\test\twitter\image'
    path = os.path.join(dir_name,file_name)
    if not os.path.exists(path):
        response = urllib.request.urlopen(url)
        if response.status == 200:
            with open(path, "wb") as f:
                f.write(response.read())
                time.sleep(0.5)
```

```python
for images in tqdm(tweet_df['images']):
    for url in images:
        download(url)
```
